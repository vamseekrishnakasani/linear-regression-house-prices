```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Example data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1) 
y = np.array([2, 4, 5, 4, 5])               
```

```python
# Create and fit the linear regression model
model = LinearRegression()
model.fit(x, y)
```

```python
# Get the slope (m) and intercept (c)
m = model.coef_[0]
c = model.intercept_
print(m, c)
```

```python
# Generate predictions
y_pred = model.predict(x)
print(y_pred)
```

```python
# Plot the data points
plt.scatter(x, y, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(x, y_pred, color='red', label=f'Regression Line (y = {m:.2f}x + {c:.2f})')

# Add labels and legend
plt.xlabel('Input Variable (x)')
plt.ylabel('Output Variable (y)')
plt.title('Linear Regression with One Input Variable')
plt.legend()

# Show the plot
plt.show()
```

```python
# Actual and predicted values
y_pred = np.array([2.8, 3.4, 4.0, 4.6, 5.2])
errors = y - y_pred

# Plot actual vs. predicted values
plt.scatter(range(len(y)), y, color='blue', label='Actual')
plt.scatter(range(len(y_pred)), y_pred, color='red', label='Predicted')
plt.plot(range(len(y_pred)), y_pred, color='red', linestyle='dotted', label='Regression Line')
plt.title('Actual vs. Predicted')
plt.legend()
plt.show()

# Plot residuals
plt.bar(range(len(errors)), errors, color='purple', label='Residuals (Errors)')
plt.axhline(0, color='black', linewidth=0.8, linestyle='dashed')
plt.title('Residuals')
plt.legend()
plt.show()
```

```python
# Calculate evaluation metrics
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)
```

```python
# Print metrics
print("Evaluation Metrics:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2): {r2:.2f}")
```

```python
# Train Ridge regression model
ridge_model = Ridge(alpha=1.0)  # Regularization strength alpha=1.0
ridge_model.fit(x, y)

# Predictions
y_pred_ridge = ridge_model.predict(x)
print(y_pred_ridge)
```

```python
# Evaluate the model
mae = mean_absolute_error(y, y_pred_ridge)
mse = mean_squared_error(y, y_pred_ridge)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred_ridge)

# Print metrics
print("Ridge Regression Evaluation Metrics:")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2): {r2:.2f}")

# Print model coefficients and intercept
print("\nRidge Model Parameters:")
print(f"Coefficient (Slope): {ridge_model.coef_[0]:.2f}")
print(f"Intercept: {ridge_model.intercept_:.2f}")
```
